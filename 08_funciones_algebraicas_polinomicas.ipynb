### Funciones polinómicas

Tiene la forma de $$P(x)=a_{n}x^{n} + a_{n-1}x^{n-1}+...+a_{2}x^{2}+a_{1}x + a_{1}$$

a una función que tiene esta forma se le llama polinomio de grado $n$. A los elementos $a$ los llamaremos coeficientes donde $a \in R$. 

**Por ejemplo:**

$$P(x)= 2x^{7} - x^{4} + 3x^{2} + 4$$

que es un polinomio de grado 7.


### Funciones potencia

Hay unas funciones que son un caso particular de las funciones polinómicas que son las funciones potencia, las cuales tienen la forma:


$$f(x)= x^{a}, a \in R$$ 


**Por ejemplo:**

$$f(x)= x^{2}$$

El dominio de $f(x)=x^{2}$ es $Dom_{f} = (-\infty, \infty)$. Su imagen es $Im_{f} = [0, \infty)$





```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definoms la funcion: Con un grado de polinomio 7 (Impar)
def f(x):
    return (2*x**7)-(x**4)+(3*x**2)+4

# Definimos un arr con 100 elementos desde -10 a 10
x = np.linspace(-10,10, 100)

y = f(x)

plt.plot(x,y)
plt.grid()
```

```python
# Definoms la funcion: Con un grado de polinomio 6 (Par)
def f(x):
    return (2*x**6)-(x**4)+(3*x**2)+4

# Definimos un arr con 100 elementos desde -10 a 10
x = np.linspace(-10,10, 100)

y = f(x)

plt.plot(x,y)
plt.grid()
```

```python
'''Funciones potencia'''
# Definoms la funcion potencia. Da el mismo resultado grafico
def f(x):
    return x**7


x = np.linspace(-10,10, 100)

y = f(x)

plt.plot(x,y)
plt.grid()
```

```python
# Definoms la funcion: Con un grado de polinomio 7 (Impar) Negativo
def f(x):
    return (-2*x**7)-(x**4)+(3*x**2)+4

# Definimos un arr con 100 elementos desde -10 a 10
x = np.linspace(-10,10, 100)

y = f(x)

plt.plot(x,y)
plt.grid()
```

```python
# Definoms la funcion: Con un grado de polinomio 6 (Par) Negativo
def f(x):
    return (-2*x**6)-(x**4)+(3*x**2)+4

# Definimos un arr con 100 elementos desde -10 a 10
x = np.linspace(-10,10, 100)

y = f(x)

plt.plot(x,y)
plt.grid()
```
