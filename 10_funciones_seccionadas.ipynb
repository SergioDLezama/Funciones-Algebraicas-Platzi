# Función seccionada

Son funciones que tienen diferentes valores definidos por un intervalo. Por ejemplo la función escalón de Heaviside: 

$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definimos la funcion

def H(x):
    Y = np.zeros(len(x))
    for idx,x in enumerate(x):
        if x >= 0:
            Y[idx] = 1
    return Y

N = 1000

# Definimos x
x = np.linspace(-10,10,num=N)

y = H(x)

plt.plot(x,y)
```

```python
# Valor absoluto
def f(x):

    lista = []

    for value in x:
        if value > 0:
            value = value
            lista.append(value)

        elif value < 0:
            value = value*-1
            lista.append(value)

        else:
            value = 0
            lista.append(value)
    
    return lista

N = 100
x = np.linspace(-10,10,num=N)
y = f(x)

print(y)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
```
