## Desplazamientos verticales y horizontales

Siendo $c$ una constante mayor que cero, entonces la gr√°fica:

- $y=f(x)+c$ se desplaza $c$ unidades hacia arriba.
- $y=f(x)-c$ se desplaza $c$ unidades hacia abajo.
- $y=f(x-c)$ se desplaza $c$ unidades hacia la derecha.
- $y=f(x+c)$ se desplaza $c$ unidades hacia la izquierda.

```python
# Definimos la cantidad de puntos para la grafica
N = 1000

# Definimos la funcion
def f(x):
    return (x**2)

# Definimos el valor de C
c = 0

# Definimos el valor de X
x = np.linspace(-5,5, num=N)

# Definimos Y
y = f(x)

# Definimos el plot y dos lineas
fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')

# La interseccion de la parabola esta en 0,0
plt.show()
```

## Alargamientos y compresiones

Siendo $c$ una constante mayor que cero, entonces la gr√°fica:

- $y=c \cdot f(x)$ alarga la gr√°fica verticalmente en un factor de $c$.
- $y= \frac{1}{c} \cdot f(x)$ comprime la gr√°fica verticalmente en un factor de $c$.
- $y=f(c \cdot x)$ comprime la gr√°fica horizontelmente en un factor de $c$.
- $y= f(\frac{1}{c} \cdot x )$ alarga la gr√°fica horizontelmente en un factor de $c$.

```python
# Definimos la cantidad de puntos para la grafica
N = 1000

# Definimos la funcion.
def f(x):
    return np.sin(x)

# Definimos el valor de C
c = 2

# Definimos el valor de X
x = np.linspace(-15,15, num=N)

# Definimos Y 
y = f(x) 

# Definimos el plot y dos lineas
fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
ax.set_ylim(-3,3)

# Funcion sin parametros extra
plt.show()
```

```python
# Definimos la cantidad de puntos para la grafica
N = 1000

# Definimos la funcion.
def f(x):
    return np.sin(x)

# Definimos el valor de C
c = 2

# Definimos el valor de X
x = np.linspace(-15,15, num=N)

# Definimos Y lo multiplicamos por C para alargar la grafica verticalmente
y = c*f(x) 

# Definimos el plot y dos lineas
fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
ax.set_ylim(-3,3)

# Funcion expandida verticalmente
plt.show()
```

```python
# Definimos la cantidad de puntos para la grafica
N = 1000

# Definimos la funcion.
def f(x):
    return np.sin(x)

# Definimos el valor de C
c = 2

# Definimos el valor de X
x = np.linspace(-15,15, num=N)

# Definimos Y lo multiplicamos por (1/C) para comprimir la grafica verticalmente
y = (1/c)*f(x) 

# Definimos el plot y dos lineas
fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
ax.set_ylim(-3,3)

# Funcion comprimida verticalmente
plt.show()
```

## Reflexiones

- $y=-f(x)$ refleja la gr√°fica respecto al eje x.
- $y=f(-x)$ refleja la gr√°fica respecto al eje y.

```python
# Definimos la cantidad de puntos para la grafica
N = 1000

# Definimos la funcion.
def f(x):
    return x**3

# Definimos el valor de C
c = 2

# Definimos el valor de X
x = np.linspace(-15,15, num=N)

# Definimos Y
y = f(x) 

# Definimos el plot y dos lineas
fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')

# Funcion original
plt.show()
```
