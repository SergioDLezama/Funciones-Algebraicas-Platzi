```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definimos la resolucion de la funcion
N = 1000

# Definimos X
x = np.linspace(-10,10,num=N)

# Definimos la primera funcion
def g(x):
    return x**2

y = g(x)

# Plot de la primera funcion
plt.plot(x,y)
```
